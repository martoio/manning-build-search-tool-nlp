```python
import json
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
dataFile = open('data.json')

rawData = json.load(dataFile)
dataFile.close()
```

```python
def tokenizer(inputString):
    tokens = []
    doc = nlp(inputString)
    for token in doc:
        if token.is_stop or token.is_punct:
            continue
        lemma = token.lemma_.lower().strip()
        tokens.append(lemma)
    return tokens

```

```python
for doc in rawData:
    doc['tokenized_text'] = tokenizer(doc['text'])
```

```python
newData = json.dumps(rawData, indent=4)

with open('newData.json', 'w') as outfile:
    outfile.write(newData)
  
```
